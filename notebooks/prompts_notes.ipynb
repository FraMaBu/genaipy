````python
# Test summary prompt

text = """
Lorem ipsum dolor sit amet, consectetur adipiscing elit. Quisque faucibus lorem vel sapien mollis egestas.
Quisque elementum, quam at sagittis tincidunt, tellus nunc ultrices purus, non dapibus nunc lectus volutpat felis.
Morbi euismod vestibulum tellus, sit amet gravida orci. Integer id scelerisque ligula. Vivamus ultrices sed velit
vitae fermentum. Mauris odio dolor, consectetur a ante id, auctor sagittis magna. Praesent viverra diam quam,
non finibus augue semper et. Quisque volutpat, ex ac vestibulum ultrices, massa elit pharetra urna,
vitae pretium magna tortor at metus.
"""

max_words = 75

SUMMARY_PROMPT_TEMPLATE = """
Instructions:
Your task is to generate a summary of the text sample.
Summarize the text sample provided below, delimited by triple backticks, in at most {max_words} words.

Text sample:
```{text}```

Summarized text:
"""


def build_summary_prompt(
    text: str, max_words: int = 150, template: str = SUMMARY_PROMPT_TEMPLATE
) -> str:
    '''Build basic summary prompt from template and input
    
    Parameters:
    - text: Sample to summarize
    - max_words: summary maximum number of words, by default 150.
    - template: prompt template to use, ust contain placeholders for all variables, by default SUMMARY_PROMPT_TEMPLATE.

    Returns:
    - prepared prompt
    '''
    return template.format(text=text, max_words=max_words)


print(build_summary_prompt(text=text, max_words=max_words))
````

````python
# Test style output prompt

STYLE_DRAFT_PROMPT_TEMPLATE = """
Text:
{text}

Instructions:
Transform the provided text into the provided template enclosed by triple back ticks.

Template:
```
{tpl}
```

Provide your response in {output_type}. Only return the final, filled template in your response.
"""

tpl = """
# Title

## Executive summary

## Conclusion
"""

tpl = tpl.strip()

text = "hello world"


def style_text_prompt(
    text: str, tpl: str, output_type: str = "markdown", template=STYLE_DRAFT_PROMPT_TEMPLATE
) -> str:
    '''Build basic summary prompt from template and input
    
    Parameters:
    - text (str): Text to style
    - tpl (str): Template of layout 
    - output_type (str): Desired output format. Defaults to 'markdown'.
    - template: prompt template to use, ust contain placeholders for all variables, by default STYLE_DRAFT_PROMPT_TEMPLATE.

    Returns:
    - str: prepared prompt
    '''
    return template.format(text=text, tpl=tpl, output_type=output_type)


print(style_text_prompt(text=text, tpl=tpl))
````
